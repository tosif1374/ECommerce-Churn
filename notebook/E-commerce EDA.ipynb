```python
def detect_outliers_iqr(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    outliers = df[(df[column] < lower) | (df[column] > upper)]
    return outliers

# Example:
outliers_tenure = detect_outliers_iqr(df, 'OrderAmountHikeFromlastYear')
print("Outliers in Tenure:", len(outliers_tenure))
```

```python
import pandas as pd
import numpy as np

# Initialize dictionary to store outliers
outlier_dict_iqr = {}

# Select numeric columns only
numeric_cols = df.select_dtypes(include=[np.number]).columns.drop('Churn')

# Iterate over each numeric column
for col in numeric_cols:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    
    # Detect outliers using IQR method
    outliers = df[col][(df[col] < Q1 - 1.5*IQR) | (df[col] > Q3 + 1.5*IQR)]
    
    # Store in dictionary
    outlier_dict_iqr[col] = outliers.tolist()  # convert to list for readability

# Print the dictionary
for col, values in outlier_dict_iqr.items():
    print(f"{col} - Number of Outliers: {len(values)}")
    print(values)
    print("-"*50)
```

```python

# Select numeric columns
numeric_cols = df.select_dtypes(include=[np.number]).columns.drop('Churn')

# Handle outliers
for col in numeric_cols:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    
    # Define outlier bounds
    lower_bound = Q1 - 1.5*IQR
    upper_bound = Q3 + 1.5*IQR
    
    # Replace outliers with median
    median = df[col].median()
    df[col] = np.where((df[col] < lower_bound) | (df[col] > upper_bound), median, df[col])
```
